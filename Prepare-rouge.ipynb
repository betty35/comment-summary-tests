```python
import os
import numpy as np
import pandas as pd
import time
```

```python
default_path = os.getcwd()
text_dir = os.path.join(default_path,"downloads\\opinosis\\topics")
file_names = os.listdir(text_dir) ### file names of each topic
summ_dir = os.path.join(default_path,"downloads\\opinosis\\summaries-gold")
```

```python
paths = pd.DataFrame(file_names,columns=['topic'])
```

```python
paths["summ"] = paths["topic"].apply(lambda x: x.replace(".txt.data",""))
paths
```

```python
def read_txt_file(file_dir):
    with open(file_dir, encoding='ISO-8859-1') as f:
        lines = f.readlines()
    lines = [x.strip() for x in lines] 
    return lines
```

```python
paths["text"] = paths["topic"].apply(lambda x: read_txt_file(os.path.join(text_dir,x)))
```

```python
def read_summs(file_dir):
    summs = []
    for file in os.listdir(file_dir):
        with open(os.path.join(file_dir,file), encoding='ISO-8859-1') as f:
            text = f.read().strip().replace("\n"," ")
        summs.append(text)
    return summs
```

```python
paths["summ_gold"] = paths["summ"].apply(lambda x: read_summs(os.path.join(summ_dir,x)))
```

```python
paths
```

```python
paths = paths.loc[:,['text','summ_gold']]
```

```python
paths["text_count"] = paths['text'].apply(lambda x: len(x))
paths["summ_count"] = paths['summ_gold'].apply(lambda x: len(x))
paths["avg"] = paths.apply(lambda row: int(row.text_count / row.summ_count), axis=1)
```

```python
paths
```

```python
texts = paths.text.tolist()
summs = paths.summ_gold.tolist()
avgs = paths["avg"].tolist()
```

```python
len1 = len(texts)
len1
```

```python
def list2str(mylist):
    text = ""
    for i in range(len(mylist)):
        text = text + mylist[i]+" "
    return text.strip()
```

```python
list2str(test[13:26])
```

```python
text_pair1 = []
summ_pair1 = []

for i in range(len1):
    text = texts[i]
    summ = summs[i]
    avg = avgs[i]
    for j in range(len(summ)):
        part_text = text[j*avg:(j+1)*avg]
        str_text = list2str(part_text)
        text_pair1.append(str_text)
        summ_pair1.append(summ[j])
```

```python
data = {'text':text_pair1, 'summ':summ_pair1}
result_df = pd.DataFrame(data)
result_df
```

```python
result_df.to_csv('output/test_opinosis.csv', index = False)
```
